```python
import numpy as np
import matplotlib.pyplot as plt
from SSC import SSC 
```

```python
%matplotlib widget
```

Load data and ground truth labels if any

```python
data_path = "./test/mixture-four-boxes-numCompos4-dim3-n400.asc"
labels_path = "./test/mixture-four-boxes-numCompos4-dim3-n400-predicted-labels.asc"
labels_path = None

data = np.loadtxt(data_path)
labels = np.loadtxt(labels_path) if labels_path is not None else None
```

Main part

```python
predicted_labels = SSC(
    data=data, 
    nb_clusters=4, 
    ground_truth=labels,
    Cst=1,
    OptM="Lasso",
    lmbda=0.001
)
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection="3d")

ax.scatter(data[:,0], data[:,1], data[:,2], c=predicted_labels)
```
